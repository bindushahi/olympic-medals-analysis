```python
import pandas as pd

# Load the dataset
medals = pd.read_csv('olympic_medals.csv')
medals.head()
```

```python
# Number of rows and columns
rows, columns = medals.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```

```python
#data types for all columns
print("Data types of each column:")
print(medals.dtypes)
```

```python
medals.columns
```

```python
#unique cities matches were held
print("Number of unique cities where matches were held:")
print(len(medals['Location'].unique()))

```

```python
# Total medals won by USA
usa_medals = len(medals[medals['Nationality'] == 'USA'])
print(f"Total medals won by USA: {usa_medals}")
```

```python
# Total medals for each nationality
medals_per_country = medals.groupby('Nationality')['Medal'].count()
print("Total medals per country:")
print(medals_per_country)
```
